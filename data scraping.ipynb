```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.flipkart.com/search?q=mobiles+under+50000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=1"
```

```python
r = requests.get(url)
print(r)
```

```python
soup = BeautifulSoup(r.text,"lxml")
print(soup)
```

```python
np = soup.find('a',class_= '_1LKTO3').get('href')
print(np)
```

```python
cnp = "https://www.flipkart.com" + np
print(cnp)
```

```python
url =  cnp
r = requests.get(url)
soup = BeautifulSoup(r.text,'lxml')
box = soup.find("div",class_="_1YokD2 _3Mn1Gg")
```

```python
Product_name = []
Prices = []
description = []
Reviews = []

url = "https://www.flipkart.com/search?q=mobiles+under+50000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=1"+str(1)
r = requests.get(url)
soup = BeautifulSoup(r.text,"lxml")
names = box.find_all("div", class_ = "_4rR01T")
print(names) 
```

```python
for i in names:
    name= i.text
    Product_name.append(name)
    print(Product_name)
```

```python
 print(len(Product_name))
```

```python
prices = box.find_all("div",class_="_30jeq3 _1_WHN1")
for i in prices:
 name = i.text
 Prices.append(name)
 print(Prices)   
```

```python
desc = box.find_all("ul",class_="_1xgFaf")
for i in desc:
    name = i.text
    description.append(name)
print(description)
```

```python
print(len(description))
```

```python
reviews = box.find_all("div",class_= "_3LWZlK")
for i in reviews:
    name = i.text
    Reviews.append(name)
    print(Reviews)
```

```python
print(len(Reviews))
```

```python
df = pd.DataFrame({"Product Name":Product_name,"Prices":Prices,"description":description,"Reviews":"Reviews"})
print(df)
```
